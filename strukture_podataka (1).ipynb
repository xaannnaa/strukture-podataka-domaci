```python
class Cvor:
    def __init__(self, podatak):
        self.podatak = podatak
        self.prev = None
        self.next = None


def ispisi(head):
    trenutni = head

    while trenutni:
        print(trenutni.podatak, end=" -> " if trenutni.next else "")
        trenutni = trenutni.next
    print()


def dodaj_na_kraj(head, podatak):
    novi = Cvor(podatak)

    if head is None:
        return novi
    trenutni = head
    while trenutni.next:
        trenutni = trenutni.next
    trenutni.next = novi
    novi.prev = trenutni

    return head

a = Cvor(2)
b = Cvor(45)
c = Cvor(3)
d = Cvor(1)

a.next = b
b.prev = a

b.next = c
c.prev = b

c.next = d
d.prev = c

print("Prije dodavanja:")
ispisi(a)

a = dodaj_na_kraj(a, 10)

print("Poslije dodavanja:")
ispisi(a)
```

```python
class Film:
    def __init__(self, naziv, zanr, godina, ocjena):
        self.naziv = naziv
        self.zanr = zanr
        self.godina = godina
        self.ocjena = ocjena
        self.prev = None
        self.next = None


def dodaj_na_kraj(head, naziv, zanr, godina, ocjena):
    novi = Film(naziv, zanr, godina, ocjena)

    if head is None:
        return novi
    trenutni = head
    while trenutni.next:
        trenutni = trenutni.next

    trenutni.next = novi
    novi.prev = trenutni
    return head


def prosjecna_ocjena_za_godinu(head, godina):
    suma = 0
    brojac = 0

    trenutni = head
    while trenutni:
        if trenutni.godina == godina:
            suma += trenutni.ocjena
            brojac += 1
        trenutni = trenutni.next

    if brojac == 0:
        return 0

    return suma / brojac


def stampaj_filmove_od_godine(head, min_godina):
    trenutni = head
    while trenutni:
        if trenutni.godina >= min_godina:
            print(trenutni.naziv, "-", trenutni.godina)

        trenutni = trenutni.next


def broj_filmova_po_zanru(head, zanr):
    brojac = 0
    trenutni = head

    while trenutni:
        if trenutni.zanr.lower() == zanr.lower():
            brojac += 1
        trenutni = trenutni.next
    return brojac


def stampaj_sve_filmove(head):
    trenutni = head

    while trenutni:
        print(trenutni.naziv, trenutni.zanr, trenutni.godina, trenutni.ocjena)
        trenutni = trenutni.next

lista = None

lista = dodaj_na_kraj(lista, "Titanic", "drama", 1997, 8.5)
lista = dodaj_na_kraj(lista, "Inception", "triler", 2010, 9.0)
lista = dodaj_na_kraj(lista, "Joker", "drama", 2019, 8.8)
lista = dodaj_na_kraj(lista, "Home Alone", "komedija", 1990, 7.6)
lista = dodaj_na_kraj(lista, "Avatar", "akcija", 2009, 8.1)

print("Svi filmovi:")
stampaj_sve_filmove(lista)

print("\nProsjecna ocjena filmova iz 2019:")
print(prosjecna_ocjena_za_godinu(lista, 2019))

print("\nFilmovi od 2000. godine:")
stampaj_filmove_od_godine(lista, 2000)

print("\nBroj filmova zanra drama:")
print(broj_filmova_po_zanru(lista, "drama"))
```

```python
def parnepar(S1, S2):

    pomocni = []

    while S1:

        element = S1.pop()

        if element%2==0:
            pomocni.append(element)

        else:
            S2.append(element)
    while pomocni:
        S1.append(pomocni.pop())
S1 = [3, 1, 4, 1, 2, 6]
S2 = []

print("Prije:")
print("S1 =", S1)
print("S2 =", S2)

parnepar(S1, S2)

print("\nPoslije:")
print("S1 =", S1)
print("S2 =", S2)
```

```python
class Stanica:
    def __init__(self, naziv):
        self.naziv = naziv
        self.putnici = 0
        self.next = None


def dodaj_stanicu(head, naziv):
    nova = Stanica(naziv)
    if head is None:
        nova.next = nova
        return nova
    trenutna = head

    while trenutna.next != head:
        trenutna = trenutna.next
    trenutna.next = nova
    nova.next = head
    return head


def ispisi_trasu(head):
    if head is None:
        return
    trenutna = head

    while True:
        print(trenutna.naziv, "- broj putnika:", trenutna.putnici)
        trenutna = trenutna.next

        if trenutna==head:
            break


def simulacija_tramvaja(head):
    if head is None:
        return

    trenutna = head
    broj_putnika = 0
    prvi_krug = True

    while True:
        print("\nTramvaj je stigao na stajalište:", trenutna.naziv)
        print("Trenutno putnika u tramvaju:", broj_putnika)

        ulaz = int(input("Broj putnika koji ulaze: "))
        izlaz = int(input("Broj putnika koji izlaze: "))

        if izlaz > broj_putnika:
            izlaz = broj_putnika

        broj_putnika = broj_putnika - izlaz + ulaz
        trenutna.putnici = broj_putnika

        print("Tramvaj kreće sa", trenutna.naziv, "sa", broj_putnika, "putnika.")

        trenutna = trenutna.next

        if trenutna == head:
            print("\nZavršen je jedan krug. Trenutna trasa:")
            ispisi_trasu(head)

            if broj_putnika == 0:
                print("\nTramvaj je prazan i završava smjenu.")
                break

N = int(input("Unesite broj stajališta: "))

lista = None

for i in range(N):
    naziv = input("Unesite naziv stajališta: ")
    lista = dodaj_stanicu(lista, naziv)

print("\nPočinje simulacija tramvaja broj 8.")
simulacija_tramvaja(lista)
```

Zadatak 10. b)

```python
class Turista:

    def __init__(self, ime, potrosnja):
        self.ime = ime
        self.potrosnja = potrosnja
        self.lijevo = None
        self.desno = None


class BinarnoStablo:

    def __init__(self):
        self.root = None

    def dodaj(self, node):

        if self.root is None:
            self.root = node
            return

        trenutni = self.root

        while True:

            if node.potrosnja< trenutni.potrosnja:

                if trenutni.lijevo is None:
                    trenutni.lijevo = node
                    break

                trenutni=trenutni.lijevo

            else:

                if trenutni.desno is None:
                    trenutni.desno = node
                    break

                trenutni = trenutni.desno

    def trazi(self, root, min_potrosnja):

        if root is None:
            return

        self.trazi(root.lijevo, min_potrosnja)

        if root.potrosnja > min_potrosnja:
            print(root.ime, "-", root.potrosnja)

        self.trazi(root.desno, min_potrosnja)

    def max_potrosnja(self, root):

        trenutni = root

        while trenutni.desno:
            trenutni = trenutni.desno

        return trenutni.ime

    def inorder(self, root):

        if root:

            self.inorder(root.lijevo)

            print(root.ime, "-", root.potrosnja)

            self.inorder(root.desno)


stablo = BinarnoStablo()

stablo.dodaj(Turista("Ana", 300))
stablo.dodaj(Turista("Ivan", 150))
stablo.dodaj(Turista("Maja", 500))
stablo.dodaj(Turista("Luka", 200))
stablo.dodaj(Turista("Sara", 700))


print("INORDER obilazak:")
stablo.inorder(stablo.root)


print("\nTuristi sa potrošnjom većom od 250:")
stablo.trazi(stablo.root, 250)


print("\nTurista sa najvećom potrošnjom:")
print(stablo.max_potrosnja(stablo.root))
```
